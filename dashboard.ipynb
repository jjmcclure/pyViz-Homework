## Panel Visualizations

In this section, you will copy the code for each plot type from your analysis notebook and place it into separate functions that Panel can use to create panes for the dashboard. 

These functions will convert the plot object to a Panel pane.

Be sure to include any DataFrame transformation/manipulation code required along with the plotting code.

Return a Panel pane object from each function that can be used to build the dashboard.

Note: Remove any `.show()` lines from the code. We want to return the plots instead of showing them. The Panel dashboard will then display the plots.

```python
# Define Panel Visualization Functions
def housing_units_per_year():
    """Housing Units Per Year."""
    
    # YOUR CODE HERE!
    units_year_df = (
        sfo_data['housing_units'].groupby([sfo_data.index]).mean()
    )
    return units_year_df.hvplot(kind='bar', 
                              ylim=[370000,390000],
                             title='Housing Units Per Year')

def average_gross_rent():
    """Average Gross Rent in San Francisco Per Year."""
    
    # YOUR CODE HERE!
    avg_rent_year = (
        sfo_data['gross_rent'].groupby([sfo_data.index]).mean()
    )
    return avg_rent_year.hvplot(kind='line',
                             title='Average Gross Rent in San Francisco Per Year')

def average_sales_price():
    """Average Sales Price Per Year."""
    
    # YOUR CODE HERE!
    avg_sale_year = (
        sfo_data['sale_price_sqr_foot'].groupby([sfo_data.index]).mean()
    )
    return avg_sale_year.hvplot(kind='line',
                             title='Average Sales Price Per Year')

def average_price_by_neighborhood():
    """Average Prices by Neighborhood."""
    
    # YOUR CODE HERE!
    year_neighborhood_df = (
        sfo_data[['neighborhood','sale_price_sqr_foot']]
        .groupby([sfo_data.index,'neighborhood'])
        .mean()
    )
    return year_neighborhood_df.hvplot(groupby='neighborhood',
                                      title='Average Prices by Neighborhood')

def top_most_expensive_neighborhoods():
    """Top 10 Most Expensive Neighborhoods."""
    
    # YOUR CODE HERE!
    neighborhood_df = (
        sfo_data[['neighborhood','sale_price_sqr_foot']]
        .groupby('neighborhood')
        .mean()
        .sort_values('sale_price_sqr_foot',ascending=False)
    )
    most_expensive =  neighborhood_df.nlargest(10, 'sale_price_sqr_foot')
    #note better code would read lowest and highest values and use for ylim params
    return most_expensive.hvplot(kind='bar', 
                                 ylim=[610,930],
                                title='Top 10 Most Expensive Neighborhoods')

def parallel_coordinates():
    """Parallel Coordinates Plot."""
    
    # YOUR CODE HERE!
    return px.parallel_coordinates(sfo_data.groupby(['neighborhood']).mean().reset_index().head(10),
                                  title='Parallel Coordinates Plot')

def parallel_categories():
    """Parallel Categories Plot."""
    
    # YOUR CODE HERE!
    return px.parallel_categories(sfo_data.groupby(['neighborhood']).mean().head(10),
                                 title='Parallel Categories Plot')

def neighborhood_map():
    """Neighborhood Map"""
    
    # YOUR CODE HERE!
    neighborhood_df = (
        sfo_data[['neighborhood','sale_price_sqr_foot']]
        .groupby('neighborhood')
        .mean()
        .sort_values('sale_price_sqr_foot',ascending=False)
    )
    loc_sales = pd.concat([neighborhood_df,df_neighborhood_locations], axis=1, sort=True)
    # loc_sales.isnull().count
    loc_sales.dropna(inplace=True)
    loc_sales.reset_index()
    
    #map code here
    map = px.scatter_mapbox(
        loc_sales,
        lat="Lat",
        lon="Lon",
        size="sale_price_sqr_foot",
        color=loc_sales.index,
        zoom=10,
        title='Neighborhood Map'
    )
    return map
```

## Panel Dashboard

In this section, you will combine all of the plots into a single dashboard view using Panel. Be creative with your dashboard design!

```python
# YOUR CODE HERE!
#Row1
#create a row here
row1 = pn.Row(housing_units_per_year())
#Row2
row2 = pn.Row(average_gross_rent(), average_sales_price())
#Row3
row3 = pn.Row(average_price_by_neighborhood())
#Row4
row4 = pn.Row(top_most_expensive_neighborhoods())
#Row5
row5 = pn.Row(parallel_coordinates())
#Row6
row6 = pn.Row(parallel_categories())
#Row7
row7 = pn.Row(neighborhood_map())

#Create Columns
column1 = pn.Column(
    "# San Fran Real Estate Dashboard", row1, row2, row3, row4, row5, row6, row7
)

column2 = pn.Column(
    "# General Data", row1, row2, row5, row6
)

column3 = pn.Column(
    "# Neighborhood Data", row3, row4, row7
)

#Create Tabs
panel = pn.Tabs(
    ("All", column1), ("General", column2), ("Neighborhood", column3)
)
```
